```python
#Maximum Likelihood Estimate Function
```

```python
#Maximum Likelihood Function:
def max_lik_estimate(x, y):
    
    N, D = x.shape
    sol_1 = np.dot(x.transpose(),x)
    sol_2 = np.dot(x.transpose(),y)
    sol = np.dot(sol_1,sol_2)
    theta_ml = np.linalg.inv(sol)
    return theta_ml
```

```python
theta_ml = max_lik_estimate(x,y)
```
